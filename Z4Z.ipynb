```python
G = DirectSumGroup([4]) #defines Z/4Z
elements = G.elements()  # list of all 4 elements (tuples (x1,x2))
```

```python
vals = {}
for (x,) in G.elements():
    vals[(x,)] = Fraction(x**2,16)  # defines a function f(x,y) = xy/8
func = PolynomialFunction(G,vals)

func.explain()
print(func.degree())
func1 = func*Fraction(2,1)
print(func1.degree())
image = func1.iterated_difference([(1,)]*func1.degree())
print(image.values[(0,)])
```
