This code is dedicated to classifying all polynomial function for G = Z/4ZxZ/4Z of degree 2. 
Idea is the following, we know that polynomials of degree 1 are homomorphisms + constant, so we can assume that constant is zero and think of polynomials of degree 1 as homomorphism. Then we know that if f in polynomial of degree k, then 4f is polynomial of degree k-1. So having polynomial f of degree k-1, we can look at f/4, and this will be our candidate for polynomial of degree k, but here we have a small problem, if f had m zeroes, then each of it can be replaced by 1/4,2/4,3/4 in f/4, so we may have a lot of options for f/4 for some f.

```python
G = DirectSumGroup([4, 4]) #defines Z/4ZxZ/4Z
elements = G.elements()  # list of all 16 elements (tuples (x1,x2))
```

```python
deg1_polynomials = [] #will contain all homomorphisms
pairs = [(x, y) for x in range(4) for y in range(4)]
for (x,y) in pairs:
    vars = {}
    for (i,j) in elements:
        vars[(i,j)] = Fraction(i*x+j*y,4)
    func = PolynomialFunction(G,vars)
    deg1_polynomials.append(func)
```

```python
for fun in deg1_polynomials:
    print(fun.degree_leq(1))
```

```python
import itertools

deg2_polynomials = []
m = 1
for fun in deg1_polynomials[1:]:
    zeroes = [k for k, v in fun.values.items() if v == 0]
    n = 0
    for tup in itertools.product(range(4), repeat=len(zeroes)):
        if n%1000:
            print(m,tup)
        n += 1
        vals = {}
        t = 0
        for (i,j) in elements:
            if (i,j) in zeroes:
                vars[(i,j)] = Fraction(tup[t],4)
                t += 1 
            else:
                vars[(i,j)] = Fraction(Fraction(fun((i,j))),4)
        cand_func = PolynomialFunction(G,vars)
        if(cand_func.degree_leq(2)):
            print("HI")
            deg2_polynomials.append(cand_func)
    m += 1
```
