```python
from fractions import Fraction
from math import comb
from itertools import product
import itertools
from discpoly.group import DirectSumGroup, CyclicGroup
from discpoly.discpoly import PolynomialFunction
from discpoly.utils import standart_basis
```

```python
H = DirectSumGroup([4, 4, 4]) #defines Z/4ZxZ/4ZxZ/4Z
elements = H.elements()  
```

```python
import numpy as np

def going_through_pairs(elems,r):
    if r == 0:
        return []
    elif len(elems) == 1:
        return [elems*r]
    iter_array = []
    for i in range(r+1):
        temp = going_through_pairs(elems[:-1],r-i)
        if temp != []:
            for elem in temp:
                iter_array.append(elem + [elems[-1]]*i)
        else:
            iter_array.append([elems[-1]]*i)
    return iter_array
        


vars = {}
for (x,y,z) in elements:
    vars[(x,y,z)] = Fraction(1,4)*Fraction(z*y*x*(x-1),1*1*2)

func = PolynomialFunction(H,vars)
func.explain()
print(func.degree())

n = func.degree()+1
i = 0
array = []


for h in going_through_pairs([[1,0,0],[0,1,0],[0,0,1]],func.degree()):
    if i < n:
        array.append(func.iterated_difference(list(h))((0,0,0)))
        i += 1
    else:
        string = ""
        for elem in array:
            string += str(elem) + "  "
        print(string)
        array = [func.iterated_difference(list(h))((0,0,0))]
        i = 1
        n = n-1
        if n == 1:
            print(array[0])

        

 
```
