```python
from fractions import Fraction
from math import comb
from itertools import product
import itertools
from discpoly.group import DirectSumGroup, CyclicGroup
from discpoly.discpoly import PolynomialFunction
from discpoly.utils import standart_basis
```

```python
p = 5
H = DirectSumGroup([p**2,p**2]) 
elements = H.elements()  
```

```python
vars = {}
for (x,y) in elements:
    if (x,y) == (p**2-1,p**2-1):
        vars[(x,y)] = Fraction(1,p)
    else:
        vars[(x,y)] = 0
    

func = PolynomialFunction(H,vars)
deg = func.degree()
print(deg)
dir = (1,0)
for i in range(deg):
    func.explain()
    temp = func 
    func = func.iterated_difference([dir])
    if func.is_zero_function():
        dir = (0,1)
        func = temp
for h in itertools.combinations_with_replacement([(1,0),(0,1)],deg):
    print(func.iterated_difference(list(h))((0,0)))
```

```python
vars = {}
for (x,y) in elements:
    vars[(x,y)] = Fraction(1,p*2)*Fraction(x*y,1)

    

func = PolynomialFunction(H,vars)
deg = func.degree()
print(deg)

for h in itertools.combinations_with_replacement([(1,0),(0,1)],deg):
    print(func.iterated_difference(list(h))((0,0)))
```
